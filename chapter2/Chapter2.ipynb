```python
import numpy as np

a = np.array([2, 4, 5, 1, 3])
a.sort()
print(a)
```

```python
a = np.random.randint(1, 11, 15).reshape(3, 5)
print(a)
a.sort()
print(a)
print(a.shape)
```

```python
a = np.random.randint(1, 11, 10)
print(a)
print('average = {}'.format(np.average(a)))
print('variance = {}'.format(np.var(a)))
print('standard deviatioon = {}'.format(np.std(a)))
```

```python
import matplotlib.pyplot as plt

plt.plot(a)
plt.show()
```

```python
a = np.random.randint(50, 101, 100)
print('avg = {}, std = {}'.format(np.average(a), np.std(a)))

plt.plot(a)
plt.grid()
plt.show()

plt.hist(a)
plt.grid()
plt.show()
```
